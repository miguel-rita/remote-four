```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

train_df, test_df = pd.read_csv('../data/train.csv'), pd.read_csv('../data/test.csv')
```

```python
train_df.head()
```

```python
train_df = train_df.loc[:, train_df.columns[:10]]
```

```python
train_df.head()
```

```python
fig, axs = plt.subplots(train_df.shape[1]-2, 1, figsize=(14,2 * (train_df.shape[1]-2)))
for i, col in enumerate(train_df.columns[2:]):
    #sns.distplot(x=col, y='y_tgt_r', hue='y_tgt', data=train_feats, ax=axs[i], s=2, marker='|')
    feats_0 = train_df.loc[train_df['target'] == 0, col]
    feats_1 = train_df.loc[train_df['target'] == 1, col]
    sns.distplot(a=feats_0, kde=False, bins=np.linspace(feats_0.min(), feats_0.max(), 10000), ax=axs[i])
    sns.distplot(a=feats_1, kde=False, bins=np.linspace(feats_1.min(), feats_1.max(), 10000), ax=axs[i])
    axs[i].set_xlim(7,7.5)
fig.tight_layout()
```
